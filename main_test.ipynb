```python
from src.DDO import DDO
from src.Optimizer import Optimizer
from src.Scatter import Scatter
from sklearn.gaussian_process.kernels import ConstantKernel, Matern, RBF
import numpy as np
```

```python
proj_name = "231006"
excel_name = "lhs120_init_1.xlsx"
# test_NGV
# LHS180_t3
ddo = DDO(proj_name)
ddo.read_excel(file_name=excel_name, train_ratio=0.8, coef_outlier=2)
```

```python
kernel = ConstantKernel() * Matern(length_scale=[1.]*ddo.n_var, nu=2.5, length_scale_bounds=(1e-5, 1e+7))
ddo.fit(kernel=kernel, n_restarts_optimizer=10, random_state=42, normalize_y=True)
```

```python
plot_  = Scatter(ddo)
plot_.do(save_fig=True)
```

```python
#231006 for sensitivity analysis
x_SA = np.ones((101, ddo.x.shape[1])) * 0.5
x_SA[:,3] = np.linspace(0,1,101) # DV4 sweep  SA
y_SA = ddo.predict(x_SA)
import matplotlib.pyplot as plt
plt.plot(x_SA[:,3],y_SA[:,0]) # DV4 sweep  1 QoI 
plt.show()
plt.plot(x_SA[:,3],y_SA[:,8]) # DV4 sweep  2 QoI 
plt.show()
```
